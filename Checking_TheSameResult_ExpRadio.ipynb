```python
dff = df[['B10_Pred2_ass','B11_Pred1_ass','B12_Pred1_ass','U10_Pred2_ass','U11_Pred','U20_Pred_ass','U21_Pred2_ass','U30_Pred2_ass','U31_Pred_ass','U40_Pred_ass','U41_Pred2_ass']]
dff
```

```python
import collections
```

```python
max_class_, max_count_ = [], []
for i in range(len(dff)):
    counter=collections.Counter(dff.iloc[i,:])
    max_class = counter.most_common(3)[0][0]
    max_count = counter.most_common(3)[0][1]
    max_class_.append(max_class)
    max_count_.append(max_count)
df['Dup_clas'] = max_class_
df['Dup_count'] = max_count_
```

#จัดการเลืกเฉพาะข้อมูลที่ต้องการ

```python
df_ = df[['B10_Pred2_ass','B11_Pred1_ass','B12_Pred1_ass','U10_Pred2_ass','U11_Pred','U20_Pred_ass','U21_Pred2_ass','U30_Pred2_ass','U31_Pred_ass','U40_Pred_ass','U41_Pred2_ass','Dup_clas','Dup_count','Sub_class_New','category']]
df_.head()
```

```python
df_['index_col'] = df_.index
df_
```

```python
max = df_['Dup_clas'].tolist()
act = df_['Sub_class_New'].tolist()
ai = df_['category'].tolist()
```

```python
a_ =[]
for  i in range(len(max)):
    if act[i]==max[i]:
        a = 'Correct'
        a_.append(a)
    else:
        a = 'Incorrect'
        a_.append(a) 
df_['Result_Act']  = a_

a_ =[]
for  i in range(len(max)):
    if ai[i]==max[i]:
        a = 'Same'
        a_.append(a)
    else:
        a = 'NotSame'
        a_.append(a) 
df_['Result_AI']  = a_
```
