```python
class_ = ['AB01', 'AB02','AB03','AB04','AB05','AB06','AB07','AB081','AB082','AB083','AB09','AB10','AB11','AB12','Normal']

col_name = ['Case', 'Sub_class', 'Sub_class_name', 'Num_Case']
dff2 = pd.DataFrame(columns = col_name)

df_ = df

case = list(set(df_['P ID']))
pred_label = []
num_case = []
act_label = []

for i in case:
    cas_pred = list(df_[df_['P ID'] == i]['Sub_class'])
    cas_act = list(df_[df_['P ID'] == i]['Sub_class_name'])
    pred_label.append(cas_pred[0])
    act_label.append(cas_act[0])
    num_case.append(len(cas_pred))

dff2['Sub_class'] = pred_label
dff2['Case'] = case
dff2['Num_Case'] = num_case
dff2['Sub_class_name'] = act_label  

dff2
```
