```python
import pandas as pd
df = pd.read_csv('/home/yupaporn/CSV_file/Dataset_CCA_5Abnormality_59UI.csv')
df
```

### Split train test

```python
fold = df['fold'].tolist()
split =[]
for i in range(len(fold)):
    if fold[i] in [1,2]:
        split.append('Test')
    else:
        split.append('Train')
df['Spilt'] = split
```

```python
df.columns
```

```python
df = df[['FileName', 'Case', 'Abs Position', 'Sub Position',
       'Class', 'Sub_class', 'Path Full', 'Path Crop', 'fold', 'Spilt']]
```

### add 15AB class name

```python
dfV1 = pd.read_csv('/home/yupaporn/CSV_file/Dataset_CCA_5Abnormality_59UI_V1.csv')
df['15AB'] = dfV1['15AB']
```
