```python
train_df = pd.read_csv( '/home/yupaporn/codes/USAI/Traindf.csv')#แก้
conf_mat = pd.crosstab(train_df['Views'],train_df['Sub_class'],rownames=['5-FP Class'],colnames=['15AB Class'])
conf_mat
```

```python
view = ['FP-A', 'FP-B', 'FP-C', 'FP-D', 'FP-E']
Class15AB = ['AB01', 'AB02', 'AB03', 'AB04', 'AB05', 'AB06', 'AB07', 'AB081', 'AB082', 'AB083', 'AB09', 'AB10', 'AB11', 'AB12', 'Normal']
```

```python
p_fp_ = []
fp_ = []
AB_ = []
for i in range(15) :
    fp = [1,1,1,1,1]
    FP = list(conf_mat.iloc[:,i][0:5])
    FP_ = [a +b for a, b in zip(FP, fp)]
    C_all =  sum(FP_)
    view = ['FP-A', 'FP-B', 'FP-C', 'FP-D', 'FP-E']
    for j in range(len(FP_)):
        a = FP_[j]
        p_fp_.append(a)
        AB_.append(Class15AB[i])
        fp_.append(view[j])
```
