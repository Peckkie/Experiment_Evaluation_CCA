```python
top3 = dataframe['Top3_category'].tolist()
Act = dataframe['Sub_class_New'].tolist()
cate = dataframe['category'].tolist()
U = dataframe['U41_Pred'].tolist()
```

```python
#RediologistTop3
result = list()
re = list()
for i in range(len(U)):
    if U[i] in top3[i]:
        result.append(cate[i])
        if U[i] == 'Normal':
            re.append(0)
        else:
            re.append(1)
    elif U[i] not in top3[i]:
        result.append(U[i])
        if U[i] == 'Normal':
            re.append(0)
        else:
            re.append(1)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
act = cate
pred = result

cmat = confusion_matrix(act, pred)
print('classifier accuracy = {}%'.format((100.*np.trace(cmat))/(np.sum(cmat))))

#Marking the Confusion Matrix
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(act, pred))#performance
```
