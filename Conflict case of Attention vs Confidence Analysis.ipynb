```python
import pandas as pd
df= pd.read_csv( '/home/yupaporn/CCA_Project/IoU42_Abnormal_new2.csv') #IoU186
df
```

```python
df.columns
```

```python
df = df[['num', 'FileName',
       'sub_class', 'views', 'Path Crop', 'originalImage', 'tagName1',
       'tagName2', 'Class', 'Case', 'Abs Position', 'Sub_class', 'Path Full',
       'fold', 'AI_Pred', 'U10_Pred', 'U11_Pred', 'tagName0', 'Sub_class_New',
       'Top3_category', 'Top3_Prob', 'category', 'Prob', 'FP_category',
       'U31_Pred_ass', 'U30_Pred', 'U41_Pred', 'U21_Pred', 'U40_Pred_ass',
       'U20_Pred_ass', 'U10_Pred2_ass', 'U11_Pred2', 'U30_Pred2_ass',
       'U41_Pred2_ass', 'U31_Pred2', 'U40_Pred2', 'U20_Pred2', 'U21_Pred2_ass',
       'B10_Pred1', 'B11_Pred1_ass', 'B12_Pred1_ass', 'B10_Pred2_ass',
       'B11_Pred2', 'B12_Pred2','IoU186']]
```

```python
prob = df['Prob'].tolist()
IoU = df['IoU186'].tolist()
```

```python
a =[]
b =[]
for i in range(len(IoU)):
    
    if prob[i] > 0.5:
        a.append('high')
    else:
        a.append('low')
        
    if IoU[i] > 0.80:
        b.append('correct')
    elif IoU[i] < 0.30:
        b.append('incorrect')
    else:
        b.append('none')
```

```python
df['Confidence'] = a
df['Attention'] = b
```

```python
Attention_correct_Confidence_low = df[(df['Attention'] == 'correct')& (df['Confidence'] == 'low')]
print(len(Attention_correct_Confidence_low))
```

```python
Attention_incorrect_Confidence_high = df[(df['Attention'] == 'incorrect')&(df['Confidence'] == 'high')]
print(len(Attention_incorrect_Confidence_high))
```

# Assisted Tool

```python
AssistGP5 =['B10_Pred2_ass','B11_Pred1_ass','B12_Pred1_ass','U10_Pred2_ass','U11_Pred','U20_Pred_ass','U21_Pred2_ass','U30_Pred2_ass','U31_Pred_ass','U40_Pred_ass','U41_Pred2_ass']
AssistGP = ['GP','GP','GP','GP','GP','Res','Res','nonHB','nonHB','HB','HB']
```

```python
import numpy as np
from sklearn.metrics import classification_report,confusion_matrix
```

```python
aa = list()
per = AssistGP5
probmore=Attention_correct_Confidence_low
for j in range(len(per)):
    act = probmore['Sub_class_New'].array #category, Sub_class_New
    pred = probmore[per[j]].array
    cmat = confusion_matrix(act, pred)
    acc = (np.trace(cmat))/(np.sum(cmat))
    Recall = classification_report(act, pred).split()[-9]
    Prec = classification_report(act, pred).split()[-10] 
    #acc ai 
    pred = probmore['category'].array
    cmat = confusion_matrix(act, pred)
    acc_ai = (np.trace(cmat))/(np.sum(cmat))
    #sim
    act = probmore['category'].array
    pred = probmore[per[j]].array
    cmat = confusion_matrix(act, pred)
    sim = (np.trace(cmat))/(np.sum(cmat))
    aa.append({'per':per[j],'Level':AssistGP[j],'Attention':'correct','Confidence':'low','similarity':sim,'accuracy(AI)':acc_ai,'accuracy(expert)':acc,'Precision':Prec,'Recall':Recall,'count':len(act)})
df3 = pd.DataFrame(aa)

aa = list()
per = AssistGP5
probmore=Attention_incorrect_Confidence_high
for j in range(len(per)):
    
    act = probmore['Sub_class_New'].array #category, Sub_class_New
    pred = probmore[per[j]].array
    cmat = confusion_matrix(act, pred)
    acc = (np.trace(cmat))/(np.sum(cmat))
    Recall = classification_report(act, pred).split()[-9]
    Prec = classification_report(act, pred).split()[-10] 
    #acc ai 
    pred = probmore['category'].array
    cmat = confusion_matrix(act, pred)
    acc_ai = (np.trace(cmat))/(np.sum(cmat))
    #sim
    act = probmore['category'].array
    pred = probmore[per[j]].array
    cmat = confusion_matrix(act, pred)
    sim = (np.trace(cmat))/(np.sum(cmat))
    aa.append({'per':per[j],'Level':AssistGP[j],'Attention':'incorrect','Confidence':'high','similarity':sim,'accuracy(AI)':acc_ai,'accuracy(expert)':acc,'Precision':Prec,'Recall':Recall,'count':len(act)})
df4 = pd.DataFrame(aa)
```

```python
result= pd.concat([df3,df4], ignore_index=True)
result = result.astype({'Precision': float, 'Recall': float})
result
```
