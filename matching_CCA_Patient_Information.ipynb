```python
case =[]
filename = df['filename'].tolist()
for i in range(len(filename)):
    a = filename[i].split(' ')
    if len(a) >1:
        a1 = a[0].split('.')[0]
        a1 = a1.split('-')
        if len(a1) >1 :
            a1 = a1[0]+'.'+a1[1]
        else:
            a1 = a1[0]
        a2 = a[1].split('.')[0]
        a3 = a1+'C'+a2
        case.append(a3)# ----------------------------------------------------*
    else:
        a4 = a[0].split('_Case_')
        if len(a4) >1:
            a5 = a4[1].split('.')[0]
            a6 = a4[0]+a5 
            case.append(a6)# ----------------------------------------------------*
        else:
            a7 = a4[0].split('.')
            if len(a7) >2:
                a8 = a7[0]+'.'+a7[1]
                case.append(a8)# ----------------------------------------------------*
            else:
                a9= a7[0].split('_')
                if len(a9) >1:
                    a10 = a9[0]+'-'+a9[1]
                    case.append(a10)# ----------------------------------------------------*
                else:
                    a10 = a9[0]
                    case.append(a10)# ----------------------------------------------------*
```
