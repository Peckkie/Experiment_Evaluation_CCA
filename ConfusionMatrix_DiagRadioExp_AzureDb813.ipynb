```python
class_ = ['AB01', 'AB02','AB03','AB04','AB05','AB06','AB07','AB081','AB082','AB083','AB09','AB10','AB11','AB12','Normal']

col_name = ['Case', 'Act_label', 'Pred_label','Num_PredCase']
dff2 = pd.DataFrame(columns = col_name)

df_ = abnormal

case = list(set(df_['P ID']))
pred_label = []
num_case = []
act_label = []

for i in case:
    cas_pred = set(df_[df_['P ID'] == i]['category_BiTNet'])
    cas_act = list(df_[df_['P ID'] == i]['Sub_class'])
    pred_label.append(list(cas_pred))
    act_label.append(cas_act[0])
    num_case.append(len(cas_pred))

dff2['Pred_label'] = pred_label
dff2['Case'] = case
dff2['Num_PredCase'] = num_case
dff2['Act_label'] = act_label  

dff2
```

```python
#แปลงข้อมูล

        
Act_label = dff2['Act_label'].tolist()
Act_label_ =[]
for i in Act_label:   
    a = i.split("'")
    if len(a) ==5:
        m = a[1],a[3]
        g =[]
        for j in range(2):
            g.append(m[j])
        Act_label_.append(g)
    else:
        g =[]
        g.append(a[1])
        Act_label_.append(g)
dff2['Act_label_'] = Act_label_
```
