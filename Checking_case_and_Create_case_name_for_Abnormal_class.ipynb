```python
# complete
PathFull = Abnormal['Path Full'].tolist()
case= Abnormal['Case'].tolist()
abb = Abnormal['Sub_class_New'].tolist()

caseNew =[]
for i in range(Abnormal.shape[0]):
    a = PathFull[i].split("/")
    a1 = a[-1].split(" ")
    if a1[0] == 'FP':
        ab = abb[i] + '_' + str(case[i])
    else:
        ab = a1[0] + '_' + str(case[i])
    a = ab.split("_")
    if len(a) > 2:
        if a[1] == '1':
            caseNew.append(a[0] + '_' + a[2])
        else:
            caseNew.append(ab)
    else:
        caseNew.append(ab) 
```
