```python
#complete 
Table = list()

arange =['Aerobilia', 'Calcification' , 'Hemangioma',  'IntrahepaticDuctStones',  'LiverCyst']
SubClass =['AB13', 'AB14' , 'AB15',  'AB16',  'AB17']
for i in range(len(arange)):
    Opath = '/media/tohn/HDD/VISION_dataset/USAI/CCA_US_Dataset09122564/'+arange[i]+'/'
    os.chdir('/media/tohn/HDD/VISION_dataset/USAI/CCA_US_Dataset09122564/'+arange[i]+'/') 
    imglist = glob.glob("*.JPG")
    #print(len(imglist))
    for j in imglist :
        A = j.replace('  ',' ')
        a = A.split(' ')
        a_subclass =a[0]

        #Case
        a_case = SubClass[i]+'_'+a[0][4:6]
        #print(a_case)
        #Class
        a_class = 'Abnormal'
        #SubClass
        a_subclass = SubClass[i]
        #15AB
        AB15 = arange[i]
        #Path full images
        a_pathfull = Opath+j
        #FileName
        a_filename = j
        Table.append({'FileName':a_filename, 'Case':a_case, 'Abs Position':None, 'Sub Position': None, '15AB': AB15, 'Class':a_class, 'Sub_class':a_subclass,'Path Full':a_pathfull,'Path Crop':a_pathfull})
```

```python
df = pd.DataFrame(Table)
df
```
